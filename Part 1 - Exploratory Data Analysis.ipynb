### Part 1 ‑ Exploratory data analysis

The attached _logins.json_ file contains (simulated) timestamps of user logins in a particular
geographic location. Aggregate these login counts based on 15­minute time intervals, and
visualize and describe the resulting time series of login counts in ways that best characterize the
underlying patterns of the demand. Please report/illustrate important features of the demand,
such as daily cycles. If there are data quality issues, please report them.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
logins = pd.read_json('logins.json')
```

```python
logins.head()
```

```python
logins.info()
```

All rows are formatted in datetime. There does not seem to be any missing values from the login_time column. However, there are no other columns. Since these are user login times, we will assume that each login or row equals one user.

```python
logins['user_count'] = 1
```

```python
logins.head()
```

Now we can group the data and gain some insightful information about how many users login every 15 minutes.

```python
logins_grouped = logins.groupby(pd.Grouper(key='login_time', freq='15min')).sum()
logins_grouped.head()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(logins_grouped.user_count)
plt.xlabel('Login Time')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.title('Users Logged in 01/01/70 - 04/15/70')
plt.show()
```

Based on the graph above there is some seasonality in the data that suggests, the number of users logged in at any one time varies based on the time of day.

```python
logins_grouped.describe()
```

On average, the number of people logged in every 15 minutes is approximately 9.5 users. There is one instance that has 73 users logged in. Let's take a closer look at a single day and see if there are any login patterns.

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(logins_grouped['1970-01-01 00:00:00':'1970-01-01 23:00:00'].user_count)
plt.xlabel('Login Time')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.title('Users Logged in on January 1st, 1970')

plt.subplot(2, 2, 2)
plt.plot(logins_grouped['1970-02-20 00:00:00':'1970-02-20 23:00:00'].user_count)
plt.xlabel('Login Time')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.title('Users Logged in on February 20th, 1970')

plt.subplot(2, 2, 3)
plt.plot(logins_grouped['1970-03-01 00:00:00':'1970-03-01 23:00:00'].user_count)
plt.xlabel('Login Time')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.title('Users Logged in on March 1st, 1970')

plt.subplot(2, 2, 4)
plt.plot(logins_grouped['1970-04-10 00:00:00':'1970-04-10 23:00:00'].user_count)
plt.xlabel('Login Time')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.title('Users Logged in on April 10th, 1970')

plt.show()
```

The data shows that users tend to login late into the night in relation to our location, particularly between the hours of 11PM-3AM. The geographic location of these users suggests that it is most likely the middle of the day where they are when logging on.
