### Data description

- __city__: city this user signed up in
- __phone__: primary device for this user
- __signup_date__: date of account registration; in the form ‘YYYYMMDD’
- __last_trip_date__: the last time this user completed a trip; in the form ‘YYYYMMDD’
- __avg_dist__: the average distance in miles per trip taken in the first 30 days after signup
- __avg_rating_by_driver__: the rider’s average rating over all of their trips
- __avg_rating_of_driver__: the rider’s average rating of their drivers over all of their trips
- __surge_pct__: the percent of trips taken with surge multiplier > 1
- __avg_surge__: The average surge multiplier over all of this user’s trips
- __trips_in_first_30_days__: the number of trips this user took in the first 30 days after signing up
- __ultimate_black_user__: TRUE if the user took an Ultimate Black in their first 30 days; FALSE otherwise
- __weekday_pct__: the percent of the user’s trips occurring during a weekday

```python
# change signup_date to datetime
df.signup_date = pd.to_datetime(df.signup_date)

# change last_trip_date to datetime
df.last_trip_date = pd.to_datetime(df.last_trip_date)
```

Based on the distribution of average ratings, we can see that majority of users choose to give higher ratings to drivers. We will use the median in this case as rare, lower scores bring down the overall average of ratings.

```python
df.avg_rating_of_driver.fillna(df.avg_rating_of_driver.median(), inplace=True)
df.avg_rating_by_driver.fillna(df.avg_rating_by_driver.median(), inplace=True)
```

To determine whether a user will be active after their 6th month on the system, I took a delta of __sign_up_date__ and __last_trip_date__ to showcase how many months a user has been active since sign-up. Then I can label any user having a value greater than 6 months as active after 6 months.

However, I don't believe this is enough information to go on and to determine if a user is still active after 6 months. It would be useful to get average number of trips per month as well as number of trips in the last month of activity to get a specify the definition of active (ie. having more than 3 trips within the last 30 days regardless of sign up date).

```python
month_delta = (df.last_trip_date - df.signup_date)/np.timedelta64(1, 'M')
month_delta = round(month_delta, 0)
month_delta = month_delta.astype('int64')
df.insert(6, 'active_months_since_sign_up', month_delta)
```

```python
import datetime as dt
```

```python
# change signup_date to datetime
df.signup_date = df.signup_date.map(dt.datetime.toordinal)

# change last_trip_date to datetime
df.last_trip_date = df.last_trip_date.map(dt.datetime.toordinal)
```

```python
#create a label for users that are active after 6 months
df['active_after_6_months'] = 1

for i in range(len(df)):
    if df.active_months_since_sign_up[i] <= 5:
        df.active_after_6_months.iloc[i] = 0
    else:
        df.active_after_6_months.iloc[i] = 1 
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data=df, x='trips_in_first_30_days', hue='active_after_6_months', dodge=False)
plt.legend(bbox_to_anchor=(.90, .95))
plt.title('Users Active after 6 months based on number of trips in first 30 days')
plt.show()
```

```python
#plot heat map
plt.subplots(figsize=(15,10))
plt.title('Correlation of Features with the label')
sns.heatmap(df.corr(), vmin=-1, vmax=1, center=0, annot=True)
plt.show()
```

Based on above we can see that most of the features have a very low correlation with one another. I think it would be useful to have more information regarding the average number of trips per month. I will use a simple regression model to predict the test data.

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
reg_all = LinearRegression()
reg_all.fit(X_train, y_train)
y_pred = reg_all.predict(X_test)
reg_all.score(X_test, y_test)
```
